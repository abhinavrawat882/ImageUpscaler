```python
import cv2
import matplotlib.pyplot as plt
noOFImages=800
```

```python
### importing training data 
fileLocation="archive/t/Train/HR/"
testFileLocation="archive/t/Train/LR/"
ImageSet=[]
x=[]
y=[]
for i in range(1,6):
    a=str(i)
    l=len(a)
    for somethign in range(4-l):
        a='0'+a
    trainImg=cv2.imread(fileLocation+a+".png")
    #print(a,trainImg)
    testImg1=cv2.imread(testFileLocation+a+"x4w1.png")
    testImg2=cv2.imread(testFileLocation+a+"x4w2.png")
    testImg3=cv2.imread(testFileLocation+a+"x4w3.png")
    testImg4=cv2.imread(testFileLocation+a+"x4w4.png")
    ImageSet.append([trainImg,testImg1])
    ImageSet.append([trainImg,testImg2])
    ImageSet.append([trainImg,testImg3])
    ImageSet.append([trainImg,testImg4])
    y.append(trainImg)
    y.append(trainImg)
    y.append(trainImg)
    y.append(trainImg)
    x.append(testImg1)
    x.append(testImg2)
    x.append(testImg3)
    x.append(testImg4)
```

```python
import numpy as np
x = np.asarray(x)
```

```python
y = np.asarray(y)
```

```python
xup=[]
for i in range(len(x)):
    xup.append(cv2.resize(x[i],(y[i].shape[1],y[i].shape[0]), interpolation = cv2.INTER_LINEAR))
    
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Conv2DTranspose
```

```python
model = Sequential()
model.add(Conv2D(filters=16,kernel_size=(3,3),padding="same", activation="relu"))
#model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=8,kernel_size=(3,3),padding="same", activation="relu"))
#model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=4,kernel_size=(3,3),padding="same", activation="relu"))
#model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2DTranspose(4, (3, 3), strides=2, padding="same"))
#model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
#model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(Conv2DTranspose(8, (3, 3), strides=2, padding="same"))
#model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
#model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(Conv2DTranspose(16, (3, 3), strides=2, padding="same"))
#model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(Conv2D(filters=3,kernel_size=(3,3),padding="same", activation="relu"))
```

```python
model.compile(
   optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=xup,y=y,epochs=10,batch_size=1)
```
